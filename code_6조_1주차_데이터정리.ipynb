```python
#최종
import os
import re

# 데이터셋이 있는 폴더 경로
folder_path = '/content/dataset'

for filename in os.listdir(folder_path):
    if not filename.endswith('.xlsx') or '정제완료' in filename:
        continue

    # --- [1단계: 연도 확인하고 변경하기] ---
    match = re.search(r'\d{4}', filename)
    if not match:
        continue

    old_year = match.group()

    if '전임교원' in filename or '강의료' in filename:
        new_year = old_year
    else:
        new_year = str(int(old_year) - 1)

    # --- [2단계: 지저분한 꼬리표 떼어내기 (이름 청소)] ---
    if '. ' in filename:
        clean_name = filename.split('. ')[1]
    else:
        clean_name = filename

    # ✨ [핵심 보완 포인트] 원래 이름에 남아있는 '2023년_', '2023_' 같은 옛날 연도를 싹 지워줍니다!
    clean_name = re.sub(r'^\d{4}년?_?', '', clean_name)

    # 나머지 불필요한 글자들 지우기
    clean_name = clean_name.replace('_학교별자료', '')
    clean_name = clean_name.replace(' (1)', '')
    clean_name = clean_name.replace('.xlsx', '')
    clean_name = clean_name.strip()

    # --- [3단계: 새 이름표 붙여주기] ---
    new_filename = f"{new_year}_{clean_name}.xlsx"

    old_file_path = os.path.join(folder_path, filename)
    new_file_path = os.path.join(folder_path, new_filename)

    # 이름표 바꿔 달기!
    os.rename(old_file_path, new_file_path)
    print(f"완료: {filename} ➡️ {new_filename}")

print("🎉 옛날 연도를 깔끔하게 지우고 새 이름표 달기가 끝났습니다!")
```

```python
import os
import shutil

# 데이터셋이 있는 기준 폴더 경로
base_folder = '/content/dataset'

# 우리가 정한 6가지 핵심 키워드
categories = ['중도탈락', '전임교원', '장학금', '기술이전', '연구비', '강의료']

# --- [1단계: 6개의 전용 방과 1개의 임시 대기실 만들기] ---
# 6개 키워드 폴더에 '미분류' 폴더까지 추가해서 만듭니다.
for category in categories + ['미분류']:
    folder_path = os.path.join(base_folder, category)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

# --- [2단계: 파일들을 조건에 맞게 방으로 이사시키기] ---
for filename in os.listdir(base_folder):
    file_path = os.path.join(base_folder, filename)

    # 폴더는 건너뛰고, 진짜 '엑셀 파일(.xlsx)'인 경우에만 작업 시작!
    if os.path.isfile(file_path) and filename.endswith('.xlsx'):

        # '이 파일이 방을 찾았나요?' 물어보는 체크 표시 (처음엔 False)
        is_classified = False

        for category in categories:
            # 파일 이름표에 키워드가 찰떡같이 들어있는지 확인
            if category in filename:
                target_folder = os.path.join(base_folder, category)
                target_file_path = os.path.join(target_folder, filename)

                # 파일을 해당 카테고리 폴더로 쏙 이동!
                shutil.move(file_path, target_file_path)
                print(f"이사 완료: {filename} ➡️ [{category}] 방")

                # 방을 찾았으니 체크 표시를 True로 바꾸고, 다른 키워드는 안 찾아봄 (break)
                is_classified = True
                break

        # 6개 키워드를 다 찾아봤는데도 방을 못 찾은 파일이 있다면? (is_classified가 여전히 False라면)
        if not is_classified:
            target_folder = os.path.join(base_folder, '미분류')
            target_file_path = os.path.join(target_folder, filename)

            # 미분류 방으로 이동! (이렇게 해야 누락되는 파일이 1개도 없어요)
            shutil.move(file_path, target_file_path)
            print(f"임시 대기: {filename} ➡️ [미분류] 방으로 이동 (이름 확인 필요!)")

print("모든 파일의 폴더 이사 작업이 완벽하게 끝났습니다!")
```

```python
import os
import pandas as pd

# 원본 방(폴더)과 새롭게 만들 정제 방(폴더) 경로 설정
source_folder = '/content/dataset/중도탈락'
target_folder = '/content/dataset/중도탈락(정제)'

# 만약 '중도탈락(정제)' 방이 아직 없다면 새로 예쁘게 지어줍니다!
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# 1. 원본 방에 있는 파일들을 하나씩 꺼내보기
for filename in os.listdir(source_folder):
    # 엑셀 파일이 아니면 건너뛰기!
    if not filename.endswith('.xlsx'):
        continue

    year_str = filename[:4]
    if year_str.isdigit() and (2010 <= int(year_str) <= 2023):
        file_path = os.path.join(source_folder, filename)

        # [마법의 돋보기 장착!]
        df_temp = pd.read_excel(file_path, header=None)
        header_idx = 0
        for i in range(10):
            row_text = "".join([str(val).replace(' ', '') for val in df_temp.iloc[i].values])
            if '학교' in row_text and '재적' in row_text:
                header_idx = i
                break

        df = pd.read_excel(file_path, header=header_idx)
        df.columns = df.columns.astype(str).str.replace('\n', '').str.replace(' ', '')

        # [융통성 100% 기둥 찾기!]
        col_year = [c for c in df.columns if '기준' in c and ('년도' in c or '연도' in c)]
        col_school = [c for c in df.columns if c in ['학교', '학교명', '대학명', '대학']]
        col_students = [c for c in df.columns if '재적학생' in c]
        col_ratio = [c for c in df.columns if '중도탈락' in c and '비율' in c]

        # 4개의 기둥을 무사히 다 찾았다면?
        if col_year and col_school and col_students and col_ratio:
            df_extracted = df[[col_year[0], col_school[0], col_students[0], col_ratio[0]]].copy()
            df_extracted.columns = ['기준연도', '학교', '재적학생(A)', '중도탈락학생비율(%) (B/A) x 100']

            # [보너스 치료!] 엑셀 병합 때문에 비어있는 기준연도를 아래로 쭈욱 채워주기!
            df_extracted['기준연도'] = df_extracted['기준연도'].ffill()

            # ✨ [새로운 방으로 이사!] 완성된 파일을 '중도탈락(정제)' 폴더에 따로 저장하기
            new_filename = f"{year_str}_중도탈락_정제완료.xlsx"
            save_path = os.path.join(target_folder, new_filename) # <-- 목적지가 새 방으로 바뀌었어요!

            df_extracted.to_excel(save_path, index=False)
            print(f"새 방으로 이사 완료: {filename} ➡️ [{target_folder}] 안의 {new_filename}")
        else:
            print(f"앗! [{filename}] 파일은 이름표 모양이 너무 달라서 놓쳤어요.")

print("\n🎉 완벽해요! 모든 파일이 [중도탈락(정제)] 폴더에 안전하게 저장되었습니다!")
```

```python
import os
import pandas as pd

# 1. 방 준비하기
source_folder = '/content/dataset/연구비'
target_folder = '/content/dataset/연구비(정제)'

if not os.path.exists(target_folder):
    os.makedirs(target_folder)

for filename in os.listdir(source_folder):
    if not filename.endswith('.xlsx') or '정제완료' in filename:
        continue

    year_str = filename[:4]
    if year_str.isdigit() and (2010 <= int(year_str) <= 2023):
        file_path = os.path.join(source_folder, filename)

        # 엑셀 전체를 바둑판처럼 가져오기
        df = pd.read_excel(file_path, header=None)

        col_year = -1
        col_school = -1
        col_base = -1

        # 위에서부터 10줄을 훑어보면서 기둥 찾기
        for r in range(10):
            for c in range(len(df.columns)):
                # 칸 안의 띄어쓰기와 줄바꿈만 지우고 확인
                cell_val = str(df.iloc[r, c]).replace(' ', '').replace('\n', '')

                # [수정 1] 정확하게 '학교'라는 글자와 100% 일치할 때만 가져오기
                if '기준' in cell_val and ('년도' in cell_val or '연도' in cell_val):
                    col_year = c
                elif cell_val == '학교':
                    col_school = c
                elif '전임교원' in cell_val and '1인당' in cell_val and '연구비' in cell_val:
                    col_base = c

        # 4개의 기둥을 다 찾았다면 오려내기
        if col_year != -1 and col_school != -1 and col_base != -1:

            df_extracted = df.iloc[:, [col_year, col_school, col_base, col_base + 1]].copy()

            # 새 이름표 덮어쓰기
            df_extracted.columns = ['기준연도', '학교', '교내 전임교원 1인당 연구비 (천원)', '교외 전임교원 1인당 연구비 (천원)']

            # [수정 2] 시킨 것만 하기: 오직 '학교' 칸이 아예 비어있는(NaN) 줄만 삭제! (특정 텍스트 삭제 로직 제거)
            df_extracted = df_extracted.dropna(subset=['학교'])

            # 연도 빈칸 꽉꽉 채우기
            df_extracted['기준연도'] = df_extracted['기준연도'].ffill()

            # 새로운 방에 저장
            new_filename = f"{year_str}_연구비_정제완료.xlsx"
            save_path = os.path.join(target_folder, new_filename)

            df_extracted.to_excel(save_path, index=False)
            print(f"새 방으로 이사 완료: {filename} ➡️ [{target_folder}] 안의 {new_filename}")
        else:
            print(f"앗! [{filename}] 파일은 지정된 이름의 기둥을 못 찾았어요.")

print("\n작업 완료! 지시하신 조건대로만 엄격하게 정제했습니다.")
```

```python
import os
import pandas as pd

# 1. 방 준비하기
source_folder = '/content/dataset/기술이전'
target_folder = '/content/dataset/기술이전(정제)'

if not os.path.exists(target_folder):
    os.makedirs(target_folder)

for filename in os.listdir(source_folder):
    if not filename.endswith('.xlsx') or '정제완료' in filename:
        continue

    year_str = filename[:4]
    if year_str.isdigit() and (2010 <= int(year_str) <= 2023):
        file_path = os.path.join(source_folder, filename)

        # [마법의 돋보기] 진짜 이름표가 있는 줄 찾기
        df_temp = pd.read_excel(file_path, header=None)
        header_idx = 0
        for i in range(10):
            row_text = "".join([str(val).replace(' ', '') for val in df_temp.iloc[i].values])
            if '학교' in row_text and '건수' in row_text:
                header_idx = i
                break

        # 찾은 줄을 진짜 이름표로 삼아서 엑셀 다시 읽기
        df = pd.read_excel(file_path, header=header_idx)

        # 이름표 먼지(빈칸, 줄바꿈) 싹 청소하기
        df.columns = df.columns.astype(str).str.replace('\n', '').str.replace(' ', '')

        # 우리가 원하는 4개의 기둥 찾기
        col_year = [c for c in df.columns if '기준' in c and ('년도' in c or '연도' in c)]
        col_school = [c for c in df.columns if c in ['학교', '학교명']] # 엄격하게 학교, 학교명만!
        col_count = [c for c in df.columns if c == '건수']
        col_income = [c for c in df.columns if '수입료' in c]

        # 4개의 기둥을 다 찾았다면?
        if col_year and col_school and col_count and col_income:

            # 가위로 싹둑 오려내기
            df_extracted = df[[col_year[0], col_school[0], col_count[0], col_income[0]]].copy()

            # ✨ 병철님이 원하시는 새 이름표 덮어쓰기!
            df_extracted.columns = ['기준연도', '학교명', '건수', '수입료 (천원)']

            # [엄격한 청소] 오직 '학교명' 칸이 비어있는 줄만 삭제! (다른 텍스트 건드리지 않음)
            df_extracted = df_extracted.dropna(subset=['학교명'])

            # 연도 빈칸 채워주기
            df_extracted['기준연도'] = df_extracted['기준연도'].ffill()

            # 새로운 방에 개별 포장해서 저장!
            new_filename = f"{year_str}_기술이전_정제완료.xlsx"
            save_path = os.path.join(target_folder, new_filename)

            df_extracted.to_excel(save_path, index=False)
            print(f"새 방으로 이사 완료: {filename} ➡️ [{target_folder}] 안의 {new_filename}")
        else:
            print(f"앗! [{filename}] 파일은 기둥을 못 찾았어요.")

print("\n🎉 완벽해요! 기술이전 데이터를 예쁘게 오려내어 새 폴더에 저장 완료했습니다!")
```

```python
import pandas as pd
import os
import re  # ✨ 연도 숫자를 찾기 위한 마법 도구 추가!

source_dir = '/content/dataset/강의료'
target_dir = '/content/dataset/강의료(정제)'

if not os.path.exists(target_dir):
    os.makedirs(target_dir)

for filename in sorted(os.listdir(source_dir)):
    if filename.endswith(".xlsx") and not filename.startswith("~$"):

        # --- ✨ [핵심 추가] 연도 필터링 수문장 ---
        match = re.search(r'\d{4}', filename) # 파일명에서 4자리 숫자 찾기
        if not match:
            continue # 숫자가 아예 없으면 패스

        file_year = int(match.group())

        # 2010년 ~ 2023년 사이가 아니면 쿨하게 스킵!
        if not (2010 <= file_year <= 2023):
            print(f"⏭️ 연도 제외 스킵: {filename}")
            continue
        # ----------------------------------------

        print(f"🚀 처리 시작: {filename}")
        try:
            df = pd.read_excel(os.path.join(source_dir, filename), header=None)

            # 기본 좌표 설정 (학교는 F열 고정)
            col_map = {'연도': 0, '학교': 5, '등급': None, '구분': None, '단가': None}

            # 모든 열을 뒤져서 구분(주야)과 단가 위치 확보
            for i in range(len(df.columns)):
                col_str = " ".join(df.iloc[:15, i].astype(str).tolist()) # 상위 15줄 분석

                if '단가' in col_str or '지급기준' in col_str:
                    col_map['단가'] = i
                if any(k in col_str for k in ['주간', '야간']):
                    col_map['구분'] = i
                if any(k in col_str for k in ['등급', '직급', '호봉']):
                    col_map['등급'] = i

            # [핵심] 등급 열을 못 찾았다면 가상의 열을 생성하기 위한 준비
            has_grade_col = col_map['등급'] is not None

            # 필수 열(학교, 구분, 단가)만 있으면 진행
            if col_map['구분'] is None or col_map['단가'] is None:
                print(f"⚠️ {filename}: 구분 혹은 단가 열을 찾지 못했습니다. 스킵합니다.")
                continue

            # 데이터 추출
            indices = [col_map['연도'], col_map['학교']]
            if has_grade_col:
                indices.append(col_map['등급'])
            indices.extend([col_map['구분'], col_map['단가']])

            refined_df = df.iloc[:, indices].copy()

            # 컬럼 이름 부여 (등급 열 유무에 따라)
            if has_grade_col:
                refined_df.columns = ['기준년도', '학교', '등급', '구분', '단가']
            else:
                refined_df.columns = ['기준년도', '학교', '구분', '단가']
                refined_df.insert(2, '등급', '-') # 3번째 칸에 '-' 등급 강제 삽입

            # 병합 셀 채우기 및 필터링
            refined_df[['기준년도', '학교', '등급']] = refined_df[['기준년도', '학교', '등급']].ffill()
            refined_df['구분'] = refined_df['구분'].astype(str).str.strip()
            refined_df = refined_df[refined_df['구분'].str.contains('주간|야간', na=False)]

            # 단가 숫자 변환
            refined_df['단가'] = pd.to_numeric(refined_df['단가'], errors='coerce').fillna(0) * 1000

            # 저장
            output_path = os.path.join(target_dir, f"정제_{filename}")
            refined_df.to_excel(output_path, index=False)
            print(f"✅ {filename} 처리 완료 (등급 존재: {has_grade_col})")

        except Exception as e:
            print(f"❌ {filename} 오류: {e}")

print("\n✨ 2010~2023년 파일만 쏙쏙 골라서 완벽하게 처리가 완료되었습니다.")
```

```python
import pandas as pd
import os
import re

source_dir = '/content/dataset/장학금'
target_dir = '/content/dataset/장학금(정제)'

if not os.path.exists(target_dir):
    os.makedirs(target_dir)

for filename in sorted(os.listdir(source_dir)):
    if filename.endswith(".xlsx") and not filename.startswith("~$"):

        match = re.search(r'\d{4}', filename)
        if not match:
            continue

        file_year = int(match.group())
        if not (2010 <= file_year <= 2023):
            continue

        print(f"🚀 처리 시작: {filename}")
        try:
            df = pd.read_excel(os.path.join(source_dir, filename), header=None)

            col_map = {'연도': -1, '학교': -1, '재학생': -1, '1인당장학금': -1}

            for i in range(len(df.columns)):
                col_vals = [str(x).replace(' ', '').replace('\n', '') for x in df.iloc[:10, i]]
                col_str = "".join(col_vals).upper()

                if '기준' in col_str and ('년도' in col_str or '연도' in col_str):
                    col_map['연도'] = i
                elif '학교명' in col_str or '대학명' in col_str or '학교' in col_vals:
                    col_map['학교'] = i
                elif '1인당' in col_str and '장학금' in col_str:
                    col_map['1인당장학금'] = i
                elif '재학생' in col_str and '1인당' not in col_str:
                    col_map['재학생'] = i

            if all(v != -1 for v in col_map.values()):
                indices = [col_map['연도'], col_map['학교'], col_map['재학생'], col_map['1인당장학금']]
                refined_df = df.iloc[:, indices].copy()

                refined_df.columns = ['기준연도', '학교', '재학생(A)', '재학생 1인당 장학금(D=B/A)']
                refined_df['기준연도'] = refined_df['기준연도'].ffill()

                refined_df = refined_df.dropna(subset=['학교'])

                # --- ✨ 대참사 해결 포인트: 무차별 청소기 대신 핀셋 사용! ---
                # '학교', '학교명', '대학명'과 100% 똑같이 생긴 찌꺼기만 쏙 골라냅니다.
                refined_df = refined_df[~refined_df['학교'].astype(str).isin(['학교', '학교명', '대학명'])]
                # -------------------------------------------------------------

                output_path = os.path.join(target_dir, f"{file_year}_장학금_정제완료.xlsx")
                refined_df.to_excel(output_path, index=False)
                print(f"✅ 정제 완료: {file_year}_장학금_정제완료.xlsx")
            else:
                print(f"⚠️ {filename}: 필요한 기둥을 다 찾지 못했어요.")

        except Exception as e:
            print(f"❌ {filename} 오류: {e}")

print("\n🎉 전국 모든 대학교를 무사히 구출해냈습니다!")
```

```python
import pandas as pd
import numpy as np  # 결측치(NaN) 처리를 위한 필수 도구 추가
import os
import re

source_dir = '/content/dataset/전임교원'
target_dir = '/content/dataset/전임교원(정제)'

# 정제된 파일을 저장할 새 방 만들기
if not os.path.exists(target_dir):
    os.makedirs(target_dir)

# ==========================================
# 🧹 1단계: 불필요한 '연구 실적' 파일 싹 지우기 (청소)
# ==========================================
print("🗑️ [1단계] 골칫거리 '연구 실적' 파일 삭제 작전 시작!\n")
deleted_count = 0

# 폴더 안의 모든 파일을 하나씩 검사합니다.
for filename in os.listdir(source_dir):
    # 만약 파일 이름에 '연구 실적'이나 '연구실적'이 들어있다면?
    if '연구 실적' in filename or '연구실적' in filename:
        file_path = os.path.join(source_dir, filename)

        try:
            os.remove(file_path)  # ✨ 파이썬의 삭제 마법 주문!
            print(f"💥 삭제 완료: {filename}")
            deleted_count += 1
        except Exception as e:
            print(f"❌ 삭제 실패 ({filename}): {e}")

print(f"\n🎉 1단계 완료! 총 {deleted_count}개의 연구 실적 파일이 영원히 사라졌습니다!")
print("-" * 60 + "\n")

# ==========================================
# 🚀 2단계: '1인당 학생 수' 파일 정제 및 저장
# ==========================================
print("🚀 [2단계] '1인당 학생 수' 엑셀 정제 작전 시작!\n")

for filename in sorted(os.listdir(source_dir)):
    # 엑셀 파일이 아니거나 임시 파일(~)이면 무시
    if not filename.endswith(".xlsx") or filename.startswith("~$"):
        continue

    # 2010 ~ 2023년 파일만 걸러내기 (철통 방어!)
    match = re.search(r'\d{4}', filename)
    if not match:
        continue

    file_year = int(match.group())
    if not (2010 <= file_year <= 2023):
        continue

    # 다른 종류의 파일은 스킵!
    if '1인당 학생 수' not in filename:
        print(f"⏭️ 다른 종류의 파일 스킵: {filename}")
        continue

    print(f"▶️ 처리 중: {filename}")
    try:
        df = pd.read_excel(os.path.join(source_dir, filename), header=None)

        # 가로로 병합된 이름표 풀기 마법! (재학생 기준 기둥 찾기를 위함)
        header_area = df.iloc[:10, :].ffill(axis=1).ffill(axis=0)

        col_map = {'연도': -1, '학교': -1, '정원기준': -1, '재학생기준': -1}

        # 핵심 단어로 4개의 기둥 찾기
        for i in range(len(header_area.columns)):
            col_vals = [str(x).replace(' ', '').replace('\n', '') for x in header_area.iloc[:, i]]
            col_str = "".join(col_vals)

            if '기준' in col_str and ('년도' in col_str or '연도' in col_str):
                col_map['연도'] = i
            elif '학교명' in col_str or '대학명' in col_str or '학교' in col_str:
                col_map['학교'] = i
            elif '1인당' in col_str and ('편제' in col_str or '학생정원' in col_str):
                col_map['정원기준'] = i
            elif '1인당' in col_str and '재학생' in col_str:
                col_map['재학생기준'] = i

        # 4개의 기둥을 모두 무사히 찾았다면 추출 시작
        if all(v != -1 for v in col_map.values()):
            indices = [col_map['연도'], col_map['학교'], col_map['정원기준'], col_map['재학생기준']]
            refined_df = df.iloc[:, indices].copy()

            # 통일된 새 이름표 덮어쓰기
            refined_df.columns = [
                '기준연도',
                '학교',
                '학생정원 기준 전임교원 1인당 학생 수',
                '재학생 기준 전임교원 1인당 학생 수'
            ]

            # 빈칸 채우기 및 이름표 찌꺼기 청소
            refined_df['기준연도'] = refined_df['기준연도'].ffill()
            refined_df = refined_df.dropna(subset=['학교'])
            refined_df = refined_df[~refined_df['학교'].astype(str).isin(['학교', '학교명', '대학명'])]

            # --- ✨ 핵심 보완: 가짜 0을 '빈칸(NaN)'으로 증발시키기! ---
            cols_to_clean = ['학생정원 기준 전임교원 1인당 학생 수', '재학생 기준 전임교원 1인당 학생 수']
            refined_df[cols_to_clean] = refined_df[cols_to_clean].replace([0, '0', 0.0, '0.0'], np.nan)
            # -------------------------------------------------------------------

            output_path = os.path.join(target_dir, f"{file_year}_1인당학생수_정제완료.xlsx")

            # --- ✨ 핵심 보완: 엑셀 열 너비 자동 맞춤 (착시 현상 방지) ---
            with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
                refined_df.to_excel(writer, index=False)
                worksheet = writer.sheets['Sheet1']

                for i, col in enumerate(refined_df.columns):
                    # 각 칸의 글자 길이를 계산해서 넉넉하게 늘려주기
                    max_len = max(refined_df[col].astype(str).map(len).max(), len(str(col))) + 2
                    worksheet.column_dimensions[chr(65 + i)].width = max_len
            # -------------------------------------------------------------------

            print(f"✅ 정제 완료: {file_year}_1인당학생수_정제완료.xlsx")
        else:
            print(f"⚠️ {filename}: 기둥을 다 찾지 못했어요. (찾은 기둥: {col_map})")

    except Exception as e:
        print(f"❌ {filename} 오류: {e}")

print("\n🎉 모든 청소 및 정제 풀코스가 완벽하게 끝났습니다! 수고하셨습니다!")
```
